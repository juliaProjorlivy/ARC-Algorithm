```python
pcache_res = []
arc_res = []
f = open('pcache_res.txt', 'r')
for line in f:
  pcache_res.append(int(line))
f.close() 
f = open('./tests/data_30_2000.txt', 'r')
for line in f:
  arc_res.append(int(line.split()[-1]))
  if len(arc_res) == 100:
        break
f.close() 
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
arc_res = arc_res[:51]
pcache_res = pcache_res[:51]
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(20, 8)) 

# Set position of bar on X axis 
br1 = np.arange(len(arc_res)) 
br2 = [x + barWidth for x in br1] 


# Make the plot
plt.bar(br1, arc_res, color ='r', width = barWidth, 
        edgecolor ='grey', label ='ARC') 
plt.bar(br2, pcache_res, color ='g', width = barWidth, 
        edgecolor ='grey', label ='PERF') 

# Adding Xticks 
plt.xlabel('Tests', fontweight ='bold', fontsize = 15) 
plt.ylabel('Hits', fontweight ='bold', fontsize = 15) 
#plt.xticks([r + barWidth for r in range(len(arc_res))], 
        #['2015', '2016', '2017', '2018', '2019'])

plt.legend()
#plt.show() 
plt.savefig("comparison.png")
```
